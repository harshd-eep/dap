```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("facebook.csv")
print(df)
```

```python
df['Dir']=['UP' if df.loc[i,'Open']==df.loc[i,'Low'] else 'NT' for i in df.index]
df['Dir']=['DN' if df.loc[i,'Open']==df.loc[i,'High'] else df.loc[i,'Dir'] for i in df.index]
print(df['Dir'].to_string())# use to_string show the complete list
```

```python
frq=df['Dir'].value_counts()
print(frq,type(frq))
print(frq.index)
print(frq.values)
```

```python
x=frq.index
y=frq.values
plt.bar(x,y)
plt.show()
```
